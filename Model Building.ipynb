```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("pima-indians-diabetes.csv",names=["Number of times pregnant","Plasma glucose concentration a 2 hours in an oral glucose tolerance test","Diastolic blood pressure (mm Hg)","Triceps skin fold thickness (mm)","2-Hour serum insulin (mu U/ml)","Body mass index (weight in kg/(height in m)^2)","Diabetes pedigree function","Age(years)","Class variable (0 or 1)"],header=1)
df
```

```python
df=df.loc[9:].copy()
df
```

```python
df.reset_index(drop=True, inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
mean=df.mean(numeric_only=True)
df.fillna(mean,inplace=True)
df.isnull().sum()
```

```python
subset1=df.loc[0:50,["Diabetes pedigree function","Body mass index (weight in kg/(height in m)^2)","Diastolic blood pressure (mm Hg)","Plasma glucose concentration a 2 hours in an oral glucose tolerance test","Age(years)","Class variable (0 or 1)"]]
subset1
```

```python
subset2=df.loc[51:100,["Diabetes pedigree function","Body mass index (weight in kg/(height in m)^2)","Diastolic blood pressure (mm Hg)","Plasma glucose concentration a 2 hours in an oral glucose tolerance test","Age(years)","Class variable (0 or 1)"]]
subset2
```

```python
#row merge
merged_df1=pd.merge(subset1,subset2, how='outer')
merged_df1
```

```python
subset3=df.loc[df["Class variable (0 or 1)"]==1,["Body mass index (weight in kg/(height in m)^2)","Plasma glucose concentration a 2 hours in an oral glucose tolerance test","Triceps skin fold thickness (mm)"]]
subset3
```

```python
subset4=df.loc[df["Class variable (0 or 1)"]==1,["Diastolic blood pressure (mm Hg)","2-Hour serum insulin (mu U/ml)","Diabetes pedigree function","Age(years)"]]
subset4
```

```python
merged_df2=subset3.merge(subset4, left_index=True, right_index=True)
merged_df2
```

```python
X=df.iloc[:,1:8]
X
```

```python
y=df.loc[:,["Class variable (0 or 1)"]]
y
```

```python
X_train,X_test,y_train,y_test =train_test_split(X,y,train_size=0.75,test_size=0.25)
```

```python
X_train.shape
```

```python
y_train=y_train.values.ravel()
```

```python
import numpy as np
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import ConfusionMatrixDisplay as cm
```

```python
svc=make_pipeline(StandardScaler(),SVC(gamma='auto'))
```

```python
classifier2=svc.fit(X_train,y_train)
```

```python
predicted=classifier2.predict(X_test)
predicted
```

```python
confusion_matrix = cm.from_predictions(y_test, predicted)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,predicted)
```

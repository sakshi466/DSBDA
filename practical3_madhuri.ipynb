```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Scatterplot

```python
sns.scatterplot(df,x='age',y='balance',hue=df['deposit'], palette="coolwarm")
```

As per shown in above graph which indicates Duration vs Balance we can say  that when people are not deposites the money in the bank at that time at that 
time there balance is generally less than 10000.

# Voilinplot

```python
sns.catplot(df, x="loan", y="age",kind="violin")
plt.title("Age Distribution of Customers")
```

 Basically voilinplot is use to show the frequency of data. In above graph we plot graph for Loan vs Age .
 In above graph If the width of the violin plot is wider at a certain age range, it means more customers fall in that age group.And it concludes that if younger customers dominate the loan = Yes category, it might suggest younger
 people need more financial support.

# Histplot

```python
sns.histplot(df['age'], bins=20, kde=True, color='skyblue')
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution of Customers")
plt.show()
```

In above Histplot we plot graph for Age vs Count. Means here it shows that how many number of Customers are belongs 
to that perticular age group.We can conclude that most number of accout holders are belong to 25 to 45 years old.

```python
count1=df['age'].value_counts()
count1
```

# Distplot

```python
sns.displot(df, x="age", col="deposit", row="loan",facet_kws=dict(margin_titles=True))
```

In above Displot is shows that when customer has deposit or not at that time is customer having any loan or not.We can conclude that when customer has no loan at that time he has not deposit as compare to he has deposit.But when customer has loan at that time he has generally not having deposit in there bank account

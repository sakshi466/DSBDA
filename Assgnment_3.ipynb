```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
import numpy as np
```

```python
data = pd.read_csv("bank_data_set.csv")
data
```

```python
data.isnull().sum()
```

```python

sns.histplot(data["age"], bins=20, kde=True, color="orange", edgecolor="black", alpha=0.7, linewidth=1.2)
plt.title("Age Distribution", fontsize=14, fontweight="bold", color="darkred")
plt.xlabel("Age", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

print ("Inference:The histogram show that the data is normally distributed, and the people in this dataset   have mostly age  between 30-40. This histogram displays the frequency of people falling within different age ranges on the standardized test. Each bar represents a age range, and the height of the bar represents the frequency of people's age  in that range")
```

```python
sns.scatterplot(data=data, x="marital", y="balance", color="red", edgecolor="black", s=80, marker="o")
plt.title("Eduaction vs Bank Balance", fontsize=14, fontweight="bold", color="darkblue")
plt.xlabel("Eduaction", fontsize=12)
plt.ylabel("Bank Balance", fontsize=12)
plt.show()
print ("Inference:The scatterplot It is generally used to plot the relationship between variables , where an one variable is plotted on the x-axis is marital and  another  plotted on the y-axis is balance , the graph show that people who are  marride  have more bank accounts  as compare to others ")
```

```python
colors = sns.color_palette("pastel")
plt.figure(figsize=(8, 8))
plt.pie(
    data["job"].value_counts(), 
    labels=data["job"].value_counts().index, 
    autopct="%1.1f%%", 
    colors=colors
)
plt.title("Distribution of Different Jobs Among Bank Customers", fontsize=14, fontweight="bold", color="darkblue")
plt.show()
print("Inference:A pie chart helps organize and show data as a percentage of a whole ,here it show that mostly banck customers are from management field ")
```

```python

plt.figure(figsize=(8, 5))
sns.barplot(data=data, x="education", y="balance", edgecolor="black",  linewidth=1.2)
plt.title("Education Level vs. Bank Balance", fontsize=14, fontweight="bold", color="darkred")
plt.xlabel("Education Level", fontsize=12)
plt.ylabel("Average Balance", fontsize=12)
plt.show()
print("Inference:A barplot is used to visually compare the values of a numeric variable across different categories of a categorical variable, here it shows that customers who have done tertiary eduation have more back balance as compare to customers who have done other types of eduaction ")

```

```python
sns.pairplot(data)
print("Inference:A pairplot is used to visualize the pairwise relationships between multiple numeric variables within a dataset,here two colums values  are compare with each other  ")
```

```python
sns.scatterplot(data=data,x="duration",y="age")
print("Inference:The duration of an activity  does not significantly depend on age")
plt.show()
```
